# Spam Mail prediction

# Selecting Machine Learning Model

## Importing Libraries

```python
import numpy as np
import pandas as pd
```

## Importing Dataset

```python
dataset = pd.read_csv('./data/mail_data.csv')
X=dataset['Message']
y=dataset['Category']
```

```python
dataset.head()
```

## Encoding Dependent Veriables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

## Converting the text of the mail into vector form

```python
from sklearn.feature_extraction.text import TfidfVectorizer

feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')
```

```python
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
```

## Training on various models on Train set

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

classifier1 = LogisticRegression()
classifier1.fit(X_train_features, y_train)
```

### SVC

```python
from sklearn.svm import SVC

classifier3 = SVC(kernel='linear')
classifier3.fit(X_train_features, y_train)
```

### SVM (Radial basis function)

```python
from sklearn.svm import SVC

classifier4 = SVC(kernel = 'rbf')
classifier4.fit(X_train_features, y_train)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

classifier5 = RandomForestClassifier(n_estimators=50)
classifier5.fit(X_train_features, y_train)
```

## Making prediction on Test set using those models

### Logistic Regression 

```python
classifier1.score(X_test_features, y_test)
```

### SVC

```python
classifier3.score(X_test_features, y_test)
```

### SVM

```python
classifier4.score(X_test_features, y_test)
```

### Random Forest

```python
classifier5.score(X_test_features, y_test)
```
